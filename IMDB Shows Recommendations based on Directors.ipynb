#### Importing Libraries

```python
import numpy as np
import pandas as pd
```

#### Reading the dataset

```python
df = pd.read_csv('Shows.csv')
df
```

#### Deleting unnecessary columns

```python
del df['Unnamed: 0']
del df['ID']
del df['Year']
del df['Rating']
del df['Duration']
del df['Reviews']
del df['Link']
del df['Genre']

df.head()
```

#### Finding all Directors

```python
directors = []
for i in df['Director']:
    j = i.split('|')
    for k in j:
        directors.append(k.strip())
directors = list(set(directors))
print(len(directors))
```

#### One Hot Conversion

```python
one_hot = []
for i in df['Director']:
    dir = [j.strip() for j in i.split('|')]
    enc = [0 for i in range(len(directors))]
    for i in dir: 
        enc[directors.index(i)] = 1
    one_hot.append(enc)
```

#### Adding the One Hot Conversion into the DataFrame

```python
one_hot = np.array(one_hot)
for i in range(len(directors)):
    df[directors[i]] = one_hot[:,i]
df
```

#### Assuming the user input

```python
user = df[df['Name'] == 'Planet Earth'].values[0][2:]
```

#### Recommending similar movies based on exact Genre

```python
for i in df.values:
    temp = i[2:]
    if(list(user-temp).count(0) == len(directors)):
        print(i[0])
```

## Recommending Movies based on Similar Genre
### Assuming user input:

```python
user = df[df['Name'] == 'Planet Earth'].values[0][2:]
len(user)
```

### Creating a function which is used to find the modulus sum

```python
def modulus_sum(lst):
    s=0
    for i in lst:
        if(i == 1 or i == -1):
            s+=1;
    return s
```

### Creating a relation column in the dataframe
##### Lower the value of relation, higher the movie is related to the movie selected by the user

```python
rel = []
for i in df.values:
    rel.append(modulus_sum(i[2:]-user))
df['Relation'] = rel
df.head()
```

#### Top 25 recommended movies 

```python
df_movies = df.sort_values(by = 'Relation').head(25)
for i in df_movies['Name']:
    print(i)
```
