#### Importing Libraries

```python
import numpy as np
import pandas as pd
```

#### Reading the dataset

```python
df = pd.read_csv('imdb.csv')
df
```

#### Deleting unnecessary columns

```python
del df['Poster_Link']
del df["Released_Year"]
del df['Certificate']
del df['Runtime']
del df['IMDB_Rating']
del df['Meta_score']
del df['Director']
del df['Star1']
del df['Star2']
del df['Star3']
del df['Star4']
del df['No_of_Votes']
del df['Gross']
del df['Overview']
df.head()
```

#### Finding all the types of genre

```python
genre = []
for i in df['Genre']:
    j = i.split(',')
    for k in j:
        genre.append(k.strip())
genre = list(set(genre))
print(genre)
```

#### One Hot Conversion

```python
one_hot = []
for i in df['Genre']:
    gen = [j.strip() for j in i.split(',')]
    enc = [0 for i in genre]
    for i in gen: 
        enc[genre.index(i)] = 1
    one_hot.append(enc)
```

#### One Hot Conversion verification

```python
for i in range(len(df['Genre'])):
    if(one_hot[i].count(1) != len(df['Genre'][i].split(','))):
        print("Problem")
```

#### Adding the One Hot Conversion into the DataFrame

```python
one_hot = np.array(one_hot)
for i in range(len(genre)):
    df[genre[i]] = one_hot[:,i]
df
```

#### Assuming the user input

```python
user = df[df['Series_Title'] == 'Lifeboat'].values[0][2:]
```

#### Recommending similar movies based on exact Genre

```python
for i in df.values:
    temp = i[2:]
    if(list(user-temp).count(0) == 21):
        print(i[0])
```

## Recommending Movies based on Similar Genre
### Assuming user input:

```python
user = df[df['Series_Title'] == 'Lifeboat'].values[0][2:]
print(user)
```

### Creating a function which is used to find the modulus sum

```python
def modulus_sum(lst):
    s=0
    for i in lst:
        if(i == 1 or i == -1):
            s+=1;
    return s
```

### Creating a relation column in the dataframe
##### Lower the value of relation, higher the movie is related to the movie selected by the user

```python
rel = []
for i in df.values:
    rel.append(modulus_sum(i[2:]-user))
df['Relation'] = rel
df.head()
```

#### Top 25 recommended movies 

```python
df_movies = df.sort_values(by = 'Relation').head(25)
for i in df_movies['Series_Title']:
    print(i)
```
